## Try 1.  Ignore all tricks, do it the long/hard way of pouring on into the other, then poping it to get front when needed then pushing them all back to the first.  Only done when outstack is empty, so O(N) sometimes, but maybe O(log(N)) overall.

```python
class Queue2Stacks(object):
    
    def __init__(self):
        
        # Two Stacks
        self.instack = []  # main storage stack
        self.outstack = []  # temp to unload 1st into, to reverse the order
     
    def enqueue(self,element):
        
        # adding an element to the queue.
        # Just put it ontop of the 1st
        self.instack.append(element)
    
    def dequeue(self):
        
        # if stack2 has stuff, pop from that
        if self.outstack != []:
            return self.outstack.pop()
        
        # stack 2 is empty, so pop all of stack 1 onto it the pop first element of stack 2
        else:
            for _ in xrange(len(self.instack)):
                self.outstack.append(self.instack.pop())
            return self.outstack.pop()
```
